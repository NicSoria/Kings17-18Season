# Breakdown of the Sacramento Kings 2017-18 NBA Season

## Introduction:
This project aims to visualize how the Sacramento Kings performed relative to the other teams in the NBA, as well as evaluate the players on the roster who played significant minutes during the season. The stats evaluated were chosen based off of Dean Oliver's Four Factors of Basketball Success: Shooting, Turnovers, Free Throws, and Rebounding. 

The three datafiles used were sourced from basketball-reference.com: Kings Season Totals, League Team Statistics, and League standings. When evaluating the players indiviually, players who logged less than 500 minutes on the court were filtered out of the dataset to eliminate outliers due to a low sample size. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
#importing pandas, seaborn, and matplotlib for future use

```

```python
kings=pd.read_excel("KingsTotalsBBREF.xlsx")
MPFilter= (kings.MP > 500)
kings = kings[MPFilter]
kings = kings.assign(AST_to_TOV= kings['AST'] / kings['TOV'])
kings= kings.assign(PPS= (kings['PTS']-kings['FT'])/kings['FGA'])
kings= kings.assign(FTr= kings['FTA']/kings['FGA'])
```

```python
league= pd.read_excel("leagueBBREF.xlsx")
wins= pd.read_excel("wins18BBREF.xlsx")
league = league.assign(PPS= (league['PTS']-league['FT'])/league['FGA'])
league = league.assign(AST_to_TOV= league['AST'] / league['TOV'])
league= league.assign(FTr= league['FTA']/league['FGA'])
```

## League Wins over 82 game season

```python
wins_sorted=wins.sort_values("W", ascending=False)
winpal = {Abbreviation: "purple" if Abbreviation == "SAC" else "grey" for Abbreviation in wins.Abbreviation.unique()}
sns.set_context("notebook", font_scale=0.8)
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax= sns.barplot(y='Abbreviation', x='W', data=wins_sorted, palette=winpal)
```

### Scoring


```python
league_pts_sort =league.sort_values("PTS", ascending=False)
pal = {Abbreviation: "purple" if Abbreviation == "SAC" else "grey" for Abbreviation in league.Abbreviation.unique()}
ax = sns.barplot( y="Abbreviation", x="PTS", data=league, palette=pal)
ax.set(xlabel='Points Scored', ylabel='Team')
```

#### Points Contributed


```python
kingspts= kings.sort_values('PTS',ascending=False)
kingspts[['Name','FG%','3P%','FT%','eFG%','PTS']]
```

```python
kingsPoints =kings.sort_values('PTS', ascending= False)
ax= sns.barplot(y='Name',x='PTS',data=kingsPoints, color="purple")
ax.set(xlabel='Points',ylabel='Player')
plt.show()
```

```python
plt.title("Percentage of team's points")
ax= plt.pie(kingspts['PTS'],labels=kingspts['Name'], autopct='%1.1f%%')
```

### Shooting Pcts

```python
league_FG_sort = league.sort_values('FG%', ascending=False)
ax= sns.barplot( y="Abbreviation", x="FG%", data=league_FG_sort, palette=pal)
ax.set(xlabel='Field Goal%', ylabel='Team')
plt.show()
```

#### 3 Point %

```python
league_3P_pct_sort = league.sort_values('3P%', ascending=False)
ax= sns.barplot( y="Abbreviation", x="3P%", data=league_3P_pct_sort, palette=pal)
ax.set(xlabel='3PT %', ylabel='Team')
plt.show()
```

```python
kingsfg= kings.sort_values('eFG%',ascending=False)
kingsfg[['Name','eFG%','FG','FGA','FG%','3P%']]
```

### points per shot (PPS)

```python
league_PPS_sort = league.sort_values('PPS', ascending= False)
ax= sns.barplot( y="Abbreviation", x="PPS", data=league_PPS_sort, palette=pal)
ax.set(xlabel='Points Per Shot', ylabel='Team')
plt.show()
```

```python
kingsPPS =kings.sort_values('PPS', ascending= False)
ax= sns.barplot(y='Name',x='PPS',data=kingsPPS, color="purple")
ax.set(xlabel='PPS',ylabel='Player')
plt.show()
```

### Assist to Turnover Ratio

```python
leagueATR_sorted= league.sort_values('AST_to_TOV', ascending = False )
ax= sns.barplot( y="Abbreviation", x="AST_to_TOV", data=leagueATR_sorted, palette=pal)
ax.set(xlabel='AST to TOV Ratio', ylabel='Team')
plt.show()
```

```python
kingsATR= kings.sort_values('AST_to_TOV',ascending=False)
kingsATR[['Name','AST_to_TOV','AST','TOV']]
```

```python
ax=sns.barplot(y='Name',x='AST_to_TOV',data=kingsATR, color="purple")
ax.set(xlabel='AST_to_TOV',ylabel='Player')
plt.show()
```

### Free Throw ratio

```python
leagueFTr_sort = league.sort_values('FTr',ascending=False)
ax= sns.barplot( y="Abbreviation", x="FTr", data=leagueFTr_sort, palette=pal)
ax.set(xlabel='FTr', ylabel='Team')
plt.show()
```

```python
kingsFTr_sort= kings.sort_values('FTr', ascending= False)
ax= sns.barplot(y='Name', x='FTr', data=kingsFTr_sort, color='purple')
ax.set(xlabel='FTr',ylabel='Player')
plt.show()
```

Despite having a top 4 three point shooting percentage in the league, the Kings are in the bottom four for Points per Shot due to a lack of volume in 3 point attempts. Additionally a league low in Free Throw ratio indicates they're having to work significantly harder than other teams to score points. Despite having a team worst effective Field Goal rate by a significant margin, Kings rookie Frank Mason leads in the Kings in Assist to turnover ratio and is third in Free Throw rate.
